# VISUALIZAÇÃO DE DADOS

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
base = pd.read_csv("http://bit.ly/kaggletrain")
```

```python
base
```

#### Qual a relação de idade com o fator de sobrevivência?

```python
# Gráfico básico, linhas
plt.plot(base.Age)
```

```python
# Gráfico de barras
plt.bar(base.Survived, base.Age.mean())
```

```python
# Barras horizontais
plt.barh(base.Survived, base.Age.mean())
```

```python
# Gráfico de dispersão
plt.scatter(base.Survived, base.Age)
```

```python
# removendo valores nulos
base2 = base.dropna(axis=0, subset=['Age'])
plt.hist(base2.Age)
```

##### Montando a tela do gráfico

```python
# define o tamanho da imagem do gráfico
plt.figure(figsize=(20,10))
# adiciona um título ao gráfico
plt.title("Histograma da Idade dos Passageiros")
# label do eixo x
plt.xlabel('Idade')
# label do eixo y
plt.ylabel('Frequência')
plt.hist(base2.Age)
```

```python
    plt.figure(figsize=(20,10))
plt.title("Histograma da Idade dos Passageiros")
plt.xlabel('Idade')
plt.ylabel('Frequência')
# mudando a cor do gráfico
plt.hist(base2.Age, color='gray')
```

```python
# SEPARANDO OS GRÁFICOS ENTRE OS QUE SOBREVIVERAM E OS QUE NÃO SOBREVIVERAM
plt.figure(figsize=(25,10))
plt.title('Histograma de idade dos passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.hist(base2[base2.Survived==0].Age, color='red')
```

```python
plt.figure(figsize=(25,10))
plt.title('Histograma de idade dos passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.hist(base2[base2.Survived==1].Age, color='red')
```

#### VISUALIZANDO OS DOIS GRÁFICOS NA MESMA TELA

```python
plt.figure(figsize=(20,8))
plt.title('Histograma de idade dos passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.hist(base2[base2.Survived==0].Age, color='red')
plt.hist(base2[base2.Survived==1].Age, color='blue')
```

```python
# suavizando as cores
plt.figure(figsize=(20,8))
plt.title('Histograma de idade dos passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.hist(base2[base2.Survived==0].Age, color='red', alpha=0.3)
plt.hist(base2[base2.Survived==1].Age, color='blue', alpha=0.3)
```

```python
# dividindo em mais barras, no caso 30 (bins=30)
plt.figure(figsize=(20,8))
plt.title('Histograma de idade dos passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.hist(base2[base2.Survived==0].Age, color='red', alpha=0.3, bins=30)
plt.hist(base2[base2.Survived==1].Age, color='blue', alpha=0.3, bins=30)
```

```python
# coloca uma cor nos limites das barras
plt.figure(figsize=(20,8))
plt.title('Histograma de idade dos passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.hist(base2[base2.Survived==0].Age, color='red', alpha=0.3, bins=30, edgecolor='black')
plt.hist(base2[base2.Survived==1].Age, color='blue', alpha=0.3, bins=30, edgecolor='black')
```

```python
# coloca legenda no gráfico usando os parâmetro label no plt.hist e ajustando o local da legenda no plt.legend
plt.figure(figsize=(20,8))
plt.title('Histograma de idade dos passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.hist(base2[base2.Survived==0].Age, color='red', label='Não sobreviveu', alpha=0.5, bins=30, edgecolor='black')
plt.hist(base2[base2.Survived==1].Age, color='blue', label='Sobreviveu', alpha=0.5, bins=30, edgecolor='black')
plt.legend(loc='upper right')
```

```python
# salva o gráfico em um arquivo
plt.figure(figsize=(20,8))
plt.title('Histograma de idade dos passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.hist(base2[base2.Survived==0].Age, color='red', label='Não sobreviveu', alpha=0.5, bins=30, edgecolor='black')
plt.hist(base2[base2.Survived==1].Age, color='blue', label='Sobreviveu', alpha=0.5, bins=30, edgecolor='black')
plt.legend(loc='upper right')
plt.savefig(r'D:\Fotos\graficoidade.png')
```
